```python
# Set sampling rate and IMF count
SR = 8000
DUR = 2.0
SAMPLES = int(SR * DUR)
IMF_COUNT = 6

def load_audio(path, sr=SR, samples=SAMPLES):
    import pandas as pd
    import soundfile as sf
    import librosa
    x = None
    orig_sr = None
    try:
        x, orig_sr = sf.read(path)
    except Exception:
        x = pd.read_csv(path, header=None, sep=r'\s+', engine='python').iloc[:, 0].values
        orig_sr = SR
    if x.ndim > 1:
        x = x.mean(axis=1)
    if orig_sr != sr and orig_sr is not None:
        x = librosa.resample(x.astype(float), orig_sr=orig_sr, target_sr=sr)
    if len(x) < samples:
        x = np.pad(x, (0, samples - len(x)))
    else:
        x = x[:samples]
    if np.max(np.abs(x)) > 0:
        x = x / np.max(np.abs(x))
    return x

def compute_imfs(signal, max_imfs=IMF_COUNT):
    from PyEMD import EMD
    emd = EMD()
    imfs = emd.emd(signal)
    if imfs.size == 0:
        imfs = np.zeros((0, len(signal)))
    n = imfs.shape[0]
    if n >= max_imfs:
        return imfs[:max_imfs]
    pad = np.zeros((max_imfs - n, len(signal)))
    return np.vstack([imfs, pad]) if n > 0 else pad

# Test on the first file in HEALTHY_DIR
import os
healthy_files = [f for f in HEALTHY_DIR.glob("*") if f.suffix.lower() in ['.wav', '.txt', '.csv']]
print("Number of healthy files found:", len(healthy_files))

if len(healthy_files) > 0:
    x = load_audio(healthy_files[0])
    imfs = compute_imfs(x)
    print("Original signal length:", len(x))
    print("IMFs shape:", imfs.shape)
```

```python
def prepare_datasets():
    classes = {
        0: {'label': 'healthy', 'dir': HEALTHY_DIR},
        1: {'label': 'paralysis', 'dir': PARALYSIS_DIR}
    }
    files, labels = [], []

    for idx, c in classes.items():
        all_files = [f for f in c['dir'].glob("*") if f.is_file()]
        print(f"📂 Found {len(all_files)} files in {c['label'].upper()} directory")

        # ✅ Load all available files (no slicing)
        selected = all_files  # ← use full list instead of [:NUM_SIGNALS_TO_LOAD]

        for f in selected:
            files.append(str(f))
            labels.append(idx)

    all_signals = []
    for f, lab in zip(files, labels):
        try:
            x = load_audio(f)
            imfs = compute_imfs(x)
            all_signals.append({'orig': x, 'imfs': imfs, 'label': lab, 'file': f})
        except Exception as e:
            print(f"⚠️ Skipping {f} due to error: {e}")

    print(f"\n✅ Successfully loaded {len(all_signals)} total signals "
          f"({sum([1 for a in all_signals if a['label']==0])} healthy, "
          f"{sum([1 for a in all_signals if a['label']==1])} paralysis)")

    return all_signals

# Run and check counts
all_signals = prepare_datasets()
print("Total signals loaded:", len(all_signals))
```

```python
import torch
from torch import nn, optim
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

BATCH_SIZE = 32
CNN_EPOCHS = 15
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {DEVICE}")

class WaveDataset(Dataset):
    def __init__(self, X, y):
        self.X, self.y = X, y
    def __len__(self):
        return len(self.y)
    def __getitem__(self, idx):
        return torch.from_numpy(self.X[idx]).unsqueeze(0), torch.tensor(self.y[idx], dtype=torch.long)

class SimpleCNN1D(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(1, 16, kernel_size=9, padding=4), nn.ReLU(), nn.MaxPool1d(4),
            nn.Conv1d(16, 32, kernel_size=7, padding=3), nn.ReLU(), nn.MaxPool1d(4),
            nn.Conv1d(32, 64, kernel_size=5, padding=2), nn.ReLU(),
            nn.AdaptiveAvgPool1d(1), nn.Flatten(),
            nn.Linear(64, 32), nn.ReLU(), nn.Linear(32, 2)
        )
    def forward(self, x):
        return self.net(x)

def train_evaluate_cnn(X_train, y_train, X_test, y_test):
    model = SimpleCNN1D().to(DEVICE)
    opt = optim.Adam(model.parameters(), lr=1e-3)
    loss_fn = nn.CrossEntropyLoss()
    loader = DataLoader(WaveDataset(X_train, y_train), batch_size=BATCH_SIZE, shuffle=True)
    for epoch in range(CNN_EPOCHS):
        model.train()
        for xb, yb in loader:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            opt.zero_grad()
            loss = loss_fn(model(xb), yb)
            loss.backward()
            opt.step()
    # Evaluation
    model.eval()
    preds = []
    with torch.no_grad():
        for xb, _ in DataLoader(WaveDataset(X_test, y_test), batch_size=BATCH_SIZE):
            p = model(xb.to(DEVICE)).argmax(1).cpu().numpy()
            preds.extend(p)
    acc = accuracy_score(y_test, preds)
    return acc
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

RL_EPOCHS = 200

class PolicyNet(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 256), nn.ReLU(),
            nn.Linear(256, 64), nn.ReLU(),
            nn.Linear(64, 2), nn.Softmax(dim=-1)
        )
    def forward(self, x):
        return self.net(x)

def train_evaluate_reinforce(X_train, y_train, X_test, y_test):
    L = X_train.shape[1]
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    policy = PolicyNet(L).to(DEVICE)
    opt = optim.Adam(policy.parameters(), lr=1e-3)
    for _ in range(RL_EPOCHS):
        idxs = np.random.permutation(len(X_train))
        for i in range(0, len(idxs), 32):
            b = idxs[i:i+32]
            xb = torch.tensor(X_train[b], dtype=torch.float32).to(DEVICE)
            yb = torch.tensor(y_train[b], dtype=torch.long).to(DEVICE)
            probs = policy(xb)
            m = torch.distributions.Categorical(probs)
            acts = m.sample()
            rewards = (acts == yb).float()
            loss = -(m.log_prob(acts) * rewards).mean()
            opt.zero_grad()
            loss.backward()
            opt.step()
    with torch.no_grad():
        X_test_tensor = torch.tensor(X_test, dtype=torch.float32).to(DEVICE)
        preds = policy(X_test_tensor).argmax(1).cpu().numpy()
    acc = accuracy_score(y_test, preds)
    return acc
```

```python
from sklearn.preprocessing import StandardScaler

ELM_HIDDEN = 1000

class ELMClassifier:
    def __init__(self, n_hidden=ELM_HIDDEN):
        self.n_hidden = n_hidden
    def _act(self, x):
        return np.tanh(x)
    def fit(self, X, y):
        N, D = X.shape
        self.W = np.random.randn(D, self.n_hidden)
        self.b = np.random.randn(self.n_hidden)
        H = self._act(X.dot(self.W) + self.b)
        T = np.zeros((N, 2))
        T[np.arange(N), y] = 1
        self.Beta = np.linalg.pinv(H).dot(T)
    def predict(self, X):
        H = self._act(X.dot(self.W) + self.b)
        return (H.dot(self.Beta)).argmax(1)

def train_evaluate_elm(X_train, y_train, X_test, y_test):
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    model = ELMClassifier()
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    return acc
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd

IMF_COUNT = 6
OUTPUT_DIR = Path("./ml_results")
OUTPUT_DIR.mkdir(exist_ok=True, parents=True)

def build_feature_matrix(all_signals, signal_idx):
    X, y = [], []
    for entry in all_signals:
        sig = entry['orig'] if signal_idx == IMF_COUNT else entry['imfs'][signal_idx]
        X.append(sig.astype(np.float32))
        y.append(entry['label'])
    return np.stack(X), np.array(y)

def run_all():
    labels = [f'IMF{i+1}' for i in range(IMF_COUNT)] + ['Original']
    results = {'signal': [], 'CNN': [], 'RL': [], 'ELM': []}

    for idx, lbl in enumerate(labels):
        print(f"\n=== Running classification for {lbl} ===")
        X, y = build_feature_matrix(all_signals, idx)
        if len(X) < 4:
            print("Not enough data, skipping this IMF.")
            continue
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

        acc_cnn = train_evaluate_cnn(X_train, y_train, X_test, y_test)
        acc_rl = train_evaluate_reinforce(X_train, y_train, X_test, y_test)
        acc_elm = train_evaluate_elm(X_train, y_train, X_test, y_test)

        print(f"{lbl}: CNN={acc_cnn:.4f}, RL={acc_rl:.4f}, ELM={acc_elm:.4f}")

        results['signal'].append(lbl)
        results['CNN'].append(acc_cnn)
        results['RL'].append(acc_rl)
        results['ELM'].append(acc_elm)

    df = pd.DataFrame(results)
    df.to_csv(OUTPUT_DIR / "classification_results.csv", index=False)
    print("\n✅ Final results saved to:", OUTPUT_DIR / "classification_results.csv")
    print(df)
    return df

# Run everything
df_results = run_all()
```
